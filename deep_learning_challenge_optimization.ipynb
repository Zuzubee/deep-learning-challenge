## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'.

application_df = application_df.drop(columns = ['EIN'], axis = 1)
```

```python
# Determine the number of unique values in each column.

[(i, len(application_df[i].unique())) for i in application_df.columns]
```

```python
# Look at APPLICATION_TYPE value counts for binning

application_df.APPLICATION_TYPE.value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(application_df.APPLICATION_TYPE.value_counts()[application_df.APPLICATION_TYPE.value_counts() < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

application_df.CLASSIFICATION.value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

application_df.CLASSIFICATION.value_counts()[application_df.CLASSIFICATION.value_counts() > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(application_df.CLASSIFICATION.value_counts()[application_df.CLASSIFICATION.value_counts() < 100].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Look at NAME value counts for binning
application_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of names to be replaced
names_to_replace = list(application_df['NAME'].value_counts()[application_df['NAME'].value_counts() < 50].index)

# Replace in dataframe
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(
        name, "Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
y = application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
x_train,x_test, y_train, y_test = train_test_split(X, y, random_state = 123)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled  = x_scaler.transform(x_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

input_features = x_train_scaled.shape[1]
hidden_nodes1  = 30
hidden_nodes2  = 60
optimized_model_1 = tf.keras.models.Sequential()

# First hidden layer
optimized_model_1.add(tf.keras.layers.Dense(units=hidden_nodes1, input_dim=input_features, activation='relu'))
# Second hidden layer
optimized_model_1.add(tf.keras.layers.Dense(units=hidden_nodes2, activation='relu'))

# Output layer
optimized_model_1.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
optimized_model_1.summary()
```

```python
# Compile the model

optimized_model_1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model

fit_model_1 = optimized_model_1.fit(x_train_scaled, y_train, validation_split = 0.15, epochs = 50)
```

```python
# Evaluate the model using the test data
model_loss_1, model_accuracy_1 = optimized_model_1.evaluate(x_test_scaled, y_test, verbose = 2)
print(f"Loss: {round(model_loss_1, 3)}, Accuracy: {round(model_accuracy_1 * 100, 2)}%")
```

```python
# Export our model to HDF5 file

optimized_model_1.save('AlphabetSoupCharity_Optimization_1.h5')
```
